## Literal Typing

```python
def find_hypotenuse(x: float = 1, y: float = 2) -> float:
    """
    Calculate the hypotenuse of a triangle by given sides x, y

    args:
        - x: one side of a triangle
        - y: another side of a triangle
    return:
        length of hypotenuse
    """
    return (x ** 2 + y ** 2) ** (0.5)


print(find_hypotenuse(3, 4))


def do_something(s: str):
    return s.upper()


print(do_something('leo'))
```

### Version เต็ม

```python
# attribute: สีผม ความสูง เพศ น้ำหนัก ชื่อ 
# method: พูด ขี้ เดิน หายใจ

class Person:
    def __init__(self, name, hair_color, height, weight):
        self.name = name
        self.hair_color = hair_color
        self.height = height
        self.weight = weight
        
    def calculate_bmi(self):
        self.bmi = self.weight / ((self.height/100) ** 2)
        return self.bmi
```

$$BMI = \frac{weight(kg)}{height^2 (m.)}$$

```python
leo = Person('Leo', 'white', 171, 70)
naming = Person('Naming', 'red', 1000, 10)
print(leo.calculate_bmi())
```

```python
leo.hair_color, naming.hair_color
```

```python
class TalkingStage:
    def __init__(self, name, age, dryness, love):
        assert 0 <= love <= 10, 'love should be from 0 to 10'
        assert 0 <= dryness <= 10, 'dryness should be from 0 to 10'
        self.name = name
        self.age = age
        self.dryness = dryness
        self.love = love

    def still_likes_you(self):
        result = self.love - self.dryness
        return result

    def sai_pay(self):
        result = self.love / self.dryness
        return result
```

## Class Inheritance

```python
class Person:
    def __init__(self, age: float, height: int, nationality: str, color: str='black'):
        assert age >= 0, 'อายุมากกว่าหรือเท่ากับ 0'
        self.age = age
        self.height = height
        self.nationality = nationality
        self.color = color
    
    def speak(self, message: str) -> None:
        print(message)
```

```python
leo = Person(0.1, 0.1, 'Indian')
leo.color
leo.speak('Hello')
```

```python
class Muslim(Person):
    pass

leo = Muslim(0.1, 0.1, 'Indian')
leo.speak('อัลสามาลัยกุม')
leo.age
```

```python
class Muslim(Person):
    def __init__(self, age: float, height: int, nationality: str, color: str='black', eat_pork=False):
        super().__init__(age, height, nationality, color)
        self.eat_pork = eat_pork
    
    def lamad(self):
        print('Lamading.... >> https://www.youtube.com/watch?v=aZUXx6_GCC0&list=RDaZUXx6_GCC0&start_radio=1')

leo = Muslim(0.1, 0.1, 'Indian')
leo.lamad()
leo.eat_pork
leo.height
leo.speak('Allah')
```

```python
class Muslim(Person):
    def __init__(self, eat_pork=False, ammo_capcity=100,  **kwargs):
        super().__init__(**kwargs)
        self.eat_pork = eat_pork
        self.ammo_capacity = ammo_capcity
    
    def lamad(self):
        print('Lamading.... >> https://www.youtube.com/watch?v=aZUXx6_GCC0&list=RDaZUXx6_GCC0&start_radio=1')
    
    def equip_weapon(self, ammo):
        self.ammo_capacity -= ammo

leo = Muslim(age=0.1, height=0.1, nationality='Indian')
leo.lamad()
leo.eat_pork
leo.height
leo.speak('Allah')

leo.equip_weapon(30)
leo.equip_weapon(50)
leo.ammo_capacity
```

```python
class RectangularCoordinates:
    def __init__(self, x, y):
        self.x = x
        self.y = y
    
    def get_coordinates(self):
        return (self.x, self.y)
    
    # def is_on_axis(self):
    #     if self.x == 0 or self.y == 0:
    #         return True
    #     else:
    #         return False
        
    def is_on_axis(self):
        return self.x == 0 or self.y == 0
    
    def get_quadrant(self):
        if self.is_on_axis():
            return None
        if self.x > 0 and self.y > 0:
            return 1
        elif self.x < 0 and self.y > 0:
            return 2
        elif self.x < 0 and self.y < 0:
            return 3
        elif self.x > 0 and self.y < 0:
            return 4
        
    def distance_to_x_axis(self):
        return abs(self.y)
    
    def distance_to_y_axis(self):
        return abs(self.x)
    
    def distance_to_origin(self):
        return (self.x ** 2 + self.y ** 2) ** (0.5)
```
